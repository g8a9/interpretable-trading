```python
import pandas as pd
from glob import glob
```

```python
sec_dir = "../output/trading/sectors"
sto_dir = "../output/trading"
years = range(2007, 2018)

def read_stats(main_dir):
    return [
        pd.read_csv(glob(f"{main_dir}/out_{year}/*/classifiers_summary.csv")[0], index_col="Classifier")
        for year in years
    ]

sec_stats, sto_stats = read_stats(sec_dir), read_stats(sto_dir)
len(sec_stats), len(sto_stats)
```

```python
sto_stats = [sto.loc[sto.index.isin(sec.index.values)] for sec, sto in zip(sec_stats, sto_stats)]
```

```python
sec_stats = [s.sort_index().reset_index().set_index(["Classifier", "Year"]) for s in sec_stats]
sto_stats = [s.sort_index().reset_index().set_index(["Classifier", "Year"]) for s in sto_stats]
```

```python
sec_stats = pd.concat(sec_stats)
sto_stats = pd.concat(sto_stats)
print(sec_stats.shape, sto_stats.shape)
```

```python
var = (sec_stats - sto_stats) / sto_stats * 100
var.groupby("Year").mean().style.format("{:.2f}%")
```
